```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #to evaluating my model performanc
```

```python
data = pd.read_csv("heart.csv")
data.head()
data.tail()
```

```python
data.columns
```

```python
data.describe()
```

```python
data['target'].value_counts()  
```

1 : Defective Heart
0 : Healthy Heart

```python
# taking dependent and idependent couln target take from x axis
x = data.drop(columns='target', axis=1) #independent features
x
```

```python
y = data['target'] 
y
```

Splitting test and  train data


```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, stratify=y, random_state=2) #stratify is used to maintain the ratio of target variable in train and test data
print(x_train.shape,x_test.shape)
```

Model Training


```python
model = LogisticRegression()
```

Training with data

```python
training  = model.fit(x_train, y_train) #Training the Logistic Regression model with training data 
training.score(x_train, y_train) #accuracy on training data
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5) 
```

```python
training1 = knn_model.fit(x_train, y_train) #Training the KNN model with training data
training1.score(x_train, y_train) #accuracy on training data
```

Now Training Model Will KNN and used it

Now Predicting

```python
Y_pred = training.predict(x_test)
Y_pred#predicting the target variable on test data
```

```python
Y_train = training1.predict(x_train) #predicting the target variable on train data
print(Y_train)
```

```python
# checking accuracy
```

```python
score = accuracy_score(y_test, Y_pred ) #accuracy on train data
#error between them 
score
# So our model i working fine
```
